### __Introducción__

Matplotlib es una librería de Python que permite crear gráficos y visualizaciones de datos de una forma muy flexible y detallada. La principal herramienta dentro de Matplotlib es su módulo pyplot, que simplifica la creación de gráficos de todo tipo.

Para usar Matplotlib, primero hay que instalarlo.

_pip install matplotlib_

Una vez instalado, hay que importarlo.

```python
import matplotlib.pyplot as plt
```

### __Gráfico de línea__

Ideal para mostrar cómo cambian los datos a lo largo del tiempo o a través de algún otro eje.

```python
# Datos de ejemplo.
X = [1, 2, 3, 4, 5]
Y = [1, 4, 9, 16, 25]

# Los datos para los ejes x_values e y_values son listas de Python. Pero pueden ser arrays, diccionarios, u otros iteradores. 
# Cada valor en X corresponde a un valor en Y.
```

```python
# Crear un gráfico de línea.
plt.plot(X, Y)

# Etiquetas de los ejes.
plt.xlabel("X")
plt.ylabel("Y")

# Título del gráfico.
plt.title("Título")
plt.suptitle("Título general para toda la figura", fontsize=16)

# Tamaño de la figura.
plt.figure(figsize=(10, 5))

# Mostrar el gráfico.
plt.show()
```

Colores, formatos y funcionalidades extra.

```python
plt.plot(X, Y, color='blue', linestyle='--', marker='o')
plt.show()

# color: color de la línea.
# linestyle: tipo de línea. Puede ser "-." para línea sólida o ":" para línea de puntos.
# marker: marcador de los puntos de intersección entre X e Y. Puede ser "s", "*", "d", etc.

# Otros.
# linewidth: grosor de la línea.
# markersize: tamaño del marcador de los puntos.
# markeredgecolor: color del borde del marcador de los puntos.
# markerfacecolor: color de relleno del marcador.
# alpha: transparencia de la línea.
# label: etiqueta para el gráfico que se puede usar como leyenda.
```

Agregar etiquetas a los ejes y definir posiciones que se van a ver.

```python
plt.plot(X, Y)

# Personalizar las marcas de los ejes X e Y.
plt.xticks([1, 2, 3, 4, 5], ['Uno', 'Dos', 'Tres', 'Cuatro', 'Cinco'], rotation=45)

# locations: lista de posiciones en el eje donde se colocan las marcas.
# labels: lista de etiquetas que corresponden a cada posición.
# rotation: gira las etiquetas del eje para facilitar la lectura, útil en etiquetas largas.

plt.show()
```

Se pueden personalizar esos "ticks" o etiquetas.

```python
plt.plot(X, Y)

plt.xticks([1, 2, 3, 4, 5], ['Uno', 'Dos', 'Tres', 'Cuatro', 'Cinco'], rotation=45)

# Personalizar etiquetas.
plt.tick_params(axis='both', direction='inout', length=6, width=2, colors='red', grid_color='blue', grid_alpha=0.5)

# axis: ejes en los que aplicar los cambios ('x', 'y', o 'both').
# direction: dirección de las marcas ('in', 'out', o 'inout').
# length y width: longitud y grosor de las marcas.
# colors: color de las marcas.
# grid_color y grid_alpha: color y transparencia de la cuadrícula en el gráfico.

plt.show()
```

Definir los límites de los ejes.

```python
plt.plot(X, Y)

# Limitar el eje X de 0 a 6 y el eje Y de 0 a 30.
plt.xlim(0, 6)
plt.ylim(0, 30)

plt.show()
```

Agregar una cuadrícula al fondo del gráfico.

```python
plt.plot(X, Y)
plt.grid(True, color='gray', linestyle='--', linewidth=0.5)
plt.show()
```

Agregar una leyenda al gráfico.

```python
plt.plot(X, Y, label = "Línea azul")

plt.legend(loc='upper left', fontsize=12, frameon=True, shadow=True, facecolor='white', edgecolor='black')

# loc: define la ubicación de la leyenda (ej., 'upper left', 'lower right', 'center', etc.).
# fontsize: tamaño de la fuente de la leyenda.
# frameon: si es True, agrega un borde alrededor de la leyenda.
# shadow: agrega una sombra atrás de la leyenda.
# facecolor: color de fondo de la leyenda.
# edgecolor: color del borde de la leyenda.

plt.show()
```

Dibujar líneas horizontales o verticales.

```python
plt.plot(X, Y)

# Línea horizontal en y = 15.
plt.axhline(y=15, color='red', linestyle='--')

# Línea vertical en x = 3.
plt.axvline(x=3, color='blue', linestyle=':')

plt.show()
```

Rellenar el área entre dos líneas o entre una línea y un valor específico en el eje Y.

```python
plt.plot(X, Y)

# Rellenar el área debajo de la línea.
plt.fill_between(X, Y, color='cyan', alpha=0.2)

plt.show()
```

Agregar texto a puntos específicos del gráfico.

```python
plt.plot(X, Y)

# Forma 1.
plt.annotate('Punto Importante', xy=(3, 9), xytext=(3, 20),
             arrowprops=dict(facecolor='black', shrink=0.05))

# xy: coordenadas del punto al que se hace referencia.
# xytext: posición del texto.
# arrowprops: propiedades de la flecha, como color y tamaño.

# Forma 2.
# Añadir texto en la posición (x=3, y=10).
plt.text(2, 10, "Punto importante", fontsize=8, color='blue', ha='center')

plt.show()
```

Gráficos con subgráficos.

```python
# Crear una figura con 1 fila y 2 columnas de subgráficos.
plt.subplot(1, 2, 1)  

# Primer subgráfico.
plt.plot([1, 2, 3], [1, 4, 9], label="Cuadrado")
plt.legend()

# Segundo subgráfico.
plt.subplot(1, 2, 2)  
plt.plot([1, 2, 3], [1, 8, 27], label="Cubo")
plt.legend()

plt.show()
```

Guardar el gráfico como imagen.

```python
plt.plot(X, Y)

plt.show()
#plt.savefig("Gráfico.png", dpi=300, bbox_inches='tight')

# dpi: resolución de la imagen (dots per inch).
# bbox_inches='tight': ajusta automáticamente los márgenes para que el gráfico no tenga espacios en blanco.
```

### __Boxplot__

```python
# Datos de ejemplo.
Datos = [
    [7, 15, 13, 10, 20, 22, 23, 18, 12, 15, 8, 7, 21, 20, 18, 40],
    [8, 9, 5, 12, 15, 7, 6, 14, 9, 8, 6, 15, 9, 11, 10],
    [20, 22, 25, 24, 30, 28, 32, 35, 29, 27, 26, 33, 34, 31, 29]
]
```

```python
# Crear el boxplot.
plt.boxplot(Datos, patch_artist=True)

# patch_artist=True: rellena las cajas con color.

# Otros parámetros que se pueden agregar:
# notch=True: agrega una muesca en el boxplot para resaltar la mediana (la muesca ayuda a comparar si las medianas de diferentes grupos son significativamente diferentes).
# vert=True: orienta el boxplot de forma vertical (cambiar a False para hacerlo horizontal).
# widths=0.5: ancho de las cajas.
# boxprops: personaliza el color de la caja.
# whiskerprops: cambia el color y grosor de los bigotes.
# capprops: modifica las líneas de los extremos de los bigotes (caps).
# medianprops: ajusta el color y grosor de la mediana.
# flierprops: personaliza los outliers (color, tamaño y forma de los puntos).

plt.title("Boxplot")
plt.xlabel("Grupos de datos")
plt.ylabel("Valores")

# Etiquetas en el eje X para cada grupo.
plt.xticks([1, 2, 3], ['Grupo 1', 'Grupo 2', 'Grupo 3'])

plt.show()
```

Boxplot con todas esas opciones.

```python
plt.boxplot(Datos, patch_artist=True, notch=True, vert=True, widths=0.5, 
            boxprops=dict(facecolor="cyan", color="blue"), 
            whiskerprops=dict(color="red", linewidth=1.5),
            capprops=dict(color="green", linewidth=1.5),
            medianprops=dict(color="purple", linewidth=2),
            flierprops=dict(marker='o', markerfacecolor='orange', markersize=8, linestyle='none'))

plt.show()
```

### __Scatter plot__

Útiles para visualizar la relación entre dos variables.

```python
# Datos de ejemplo.
X = [1, 2, 3, 4, 5]
Y = [2, 4, 5, 7, 11]
```

```python
# Crear el scatter plot.
plt.scatter(X, Y)

# Añadir título y etiquetas de los ejes.
plt.title("Scatter Plot")
plt.xlabel("X")
plt.ylabel("Y")

# Mostrar el gráfico.
plt.show()
```

```python
# Tamaños de los puntos.
Tamaños = [50, 100, 200, 300, 400]  

# Colores de los puntos.
Colores = ['red', 'green', 'blue', 'orange', 'purple']  

# Crear el scatter plot.
plt.scatter(X, Y, s=Tamaños, c=Colores, alpha=0.6, edgecolor='black')

# s: tamaño de cada punto (puede ser un solo valor o una lista de valores, como en el ejemplo).
# c: color de los puntos (puede ser un color único o una lista de colores).
# alpha: transparencia de los puntos (valor entre 0 y 1).
# edgecolor: color del borde de los puntos.

# Añadir título y etiquetas de los ejes.
plt.title("Scatter Plot Personalizado")
plt.xlabel("X")
plt.ylabel("Y")

# Mostrar el gráfico.
plt.show()
```

Se puede usar un colormap (mapa de colores) para que el color de los puntos dependa de una tercera variable. Es útil si se quiere representar otra dimensión en el gráfico de dispersión.

```python
# Valores para mapear el color o tercera dimensión.
Valores = [10, 20, 30, 40, 50]  

# Crear el scatter plot.
plt.scatter(Y, Y, c=Valores, cmap='viridis', s=100, alpha=0.7, edgecolor='black')

# cmap: selecciona el colormap (por ejemplo, 'viridis', 'plasma', 'coolwarm', etc.).

# Agregar una barra de color para interpretar los valores del colormap.
plt.colorbar(label="Intensidad")

# Añadir título y etiquetas de los ejes.
plt.title("Scatter Plot con Colormap")
plt.xlabel("Valores de X")
plt.ylabel("Valores de Y")

# Mostrar el gráfico.
plt.show()
```

### __Barras__

Para comparar cantidades entre diferentes categorías.

```python
# Datos de ejemplo.
Categorias = ['A', 'B', 'C', 'D', 'E']
Valores = [5, 7, 3, 8, 6]
```

```python
# Crear el gráfico de barras.
plt.bar(Categorias, Valores)

# Título y etiquetas de los ejes.
plt.title("Gráfico de barras")
plt.xlabel("Categorías")
plt.ylabel("Valores")

# Mostrar el gráfico.
plt.show()
```

Y uno un poco más personalizado:

```python
plt.bar(Categorias, Valores, color='skyblue', edgecolor='black', width=0.5)

# color: color de las barras. Se puede usar un color único o una lista de colores.
# edgecolor: color del borde de las barras.
# width: ancho de las barras (0.5 es la mitad del ancho predeterminado).

plt.title("Gráfico de barras personalizado", pad = 15)
plt.xlabel("Categorías", labelpad = 10)
plt.ylabel("Valores", labelpad = 10)

# pad o labelpad: separación con respecto al gráfico.

plt.show()
```

Gráfico de barras horizontal.

```python
# Lo mismo, pero en vez de plt.bar es plt.hbar
plt.barh(Categorias, Valores)
plt.show()
```

Gráfico de barras apilado, o sea, una encima de la otra.

```python
# Datos de ejemplo.
Categorias = ['A', 'B', 'C', 'D', 'E']
Valores1 = [5, 7, 3, 8, 6]
Valores2 = [2, 3, 4, 1, 2]

# Crear el gráfico de barras apiladas.
plt.bar(Categorias, Valores1, color='lightblue', label='Serie 1')
plt.bar(Categorias, Valores2, bottom=Valores1, color='orange', label='Serie 2')

# Añadir título, etiquetas y leyenda.
plt.title("Gráfico de barras apiladas")
plt.xlabel("Categorías")
plt.ylabel("Valores")
plt.legend()

# Mostrar el gráfico.
plt.show()
```

Gráficos de barras agrupados, o sea, varios grupos de barras juntos.

```python
import numpy as np

# Crear una posición para cada categoría en el eje X.
X = np.arange(len(Categorias))

# Ancho de las barras.
Width = 0.4

# Crear el gráfico de barras agrupadas.
plt.bar(X - Width/2, Valores1, Width, color='lightblue', label='Serie 1')
plt.bar(X + Width/2, Valores2, Width, color='orange', label='Serie 2')
# Desplaza las barras de cada serie hacia la izquierda y derecha para que queden agrupadas.

# Añadir etiquetas a los ejes, título y leyenda.
plt.xlabel("Categorías")
plt.ylabel("Valores")
plt.title("Gráfico de barras agrupadas")
plt.xticks(X, Categorias)
plt.legend()

# Mostrar el gráfico.
plt.show()
```

### __Violín__

```python
# Generar datos de ejemplo.
Datos1 = np.random.normal(0, 1, 100)  
Datos2 = np.random.normal(2, 1.5, 100)  
Datos3 = np.random.normal(-1, 0.5, 100)  
```

```python
# Agrupar datos en una lista.
Datos = [Datos1, Datos2, Datos3]
```

```python
# Crear el gráfico de violín.
plt.violinplot(Datos, showmeans = True, showmedians = True, widths = 0.8)

# showmeans = True: muestra una línea horizontal en el centro del violín para indicar la media.
# showmedians = True: muestra una línea horizontal en el centro del violín para indicar la mediana.
# widths: controla el ancho de los violines (0.8 en este caso, para que sean más delgados).

plt.xlabel("Grupos")
plt.ylabel("Valores")
plt.show()
```

Cambiar colores al gráfico de violín.

```python
# Crear el gráfico de violín.
Violin = plt.violinplot(Datos, showmeans=True, showmedians=True)

# Cambiar el color de cada "violín".
Colores = ['skyblue', 'salmon', 'lightgreen']
for Violin, pc in enumerate(Violin['bodies']):
    pc.set_facecolor(Colores[Violin])  # Cambia el color de cada violín.
    pc.set_edgecolor('black')  # Color del borde de los violines.
    pc.set_alpha(0.7)  # Transparencia de los violines.

# Añadir título y etiquetas.
plt.xlabel("Grupos")
plt.ylabel("Valores")

# Mostrar el gráfico.
plt.show()
```
